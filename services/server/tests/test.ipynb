```python
df = pd.read_csv('../project/data/ORE/Input/Static/market_20160205.csv', header=None)
df.columns = ['Date', 'Name', 'Value']
```

```python
cpty_a_df = df[df['Name'].str.contains('CPTY_A')]
cpty_b_df = cpty_a_df.copy()
cpty_b_df['Name'] = cpty_b_df['Name'].str.replace('CPTY_A', 'CPTY_B')
cpty_b_df
```

```python
df = pd.concat([df, cpty_b_df], ignore_index=True, axis=0)
df.to_csv('../project/data/ORE/Input/Static/market_20160205_with_cpty.csv', index=False, header=False)
```
